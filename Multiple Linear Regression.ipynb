```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('house_price_sqft_living_yr_built.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
y = data['price']
x1 = data[['sqft_living','yr_built']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
